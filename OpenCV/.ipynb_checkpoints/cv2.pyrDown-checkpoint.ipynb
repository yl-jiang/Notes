```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('./anime.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots(figsize=[8, 10])
ax.imshow(img)
plt.show()
```

有两类图像金字塔：高斯金字塔和拉普拉斯金字塔

### 高斯金字塔：

<div class="alert alert-block alert-success"> 
 <b>高斯金字塔的顶部图像是通过将底部图像中的连续行和列去除得到的。顶部图像中的每一个像素值等于下一层图像中的5各像素的高斯加权平均值。这样操作一次一个MxN的图像就变成了一个(M/2)x(N/2)的图像</b>!
</div>

#### cv2.pyrDown(src[, dst[, dstsize[, borderType]]])
<div class="alert alert-block alert-success"> 
 <b>先对图像进行高斯平滑再进行下采样</b>!
</div>
+ src: 输入图像
+ dst: 目的图像(默认为输入图像的一半)
+ dstsize( src.cols/2, src.rows/2 ) :目的图像大小

```python
fig, ax = plt.subplots(figsize=[16, 20])
ax.imshow(img)
plt.show()
for i in range(4):
    img_lr = cv2.pyrDown(img)
    img = img_lr
    fig, ax = plt.subplots(figsize=[16//((i+1)*2), 20//((i+1)*2)])
    ax.imshow(img_lr)
    plt.show()
```
