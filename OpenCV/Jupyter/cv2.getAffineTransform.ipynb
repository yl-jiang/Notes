在放射变换中，原图中的所有平行线在变换后的图形中也要保持平行。为了创建这个矩阵我们需要从原图像中找到三个点以及他们在输出图像中的位置。使用cv2.getAffineTransform方法会创建一个2x3的矩阵，最后再将这个矩阵传给cv2.warpAffine应用变换。

```python
src_pts = np.float32([[50, 50], [200, 50], [50, 200]])
dst_pts = np.float32([[10, 100], [200, 50], [100, 250]])

affine_matrix = cv2.getAffineTransform(src=src_pts, dst=dst_pts)
img_affine = cv2.warpAffine(src=img, M=affine_matrix, dsize=(256, 256))

fig, ax = plt.subplots(1, figsize=[16, 10])
# ax.set_axis_off()
ax.imshow(img_affine)
plt.show()
```
