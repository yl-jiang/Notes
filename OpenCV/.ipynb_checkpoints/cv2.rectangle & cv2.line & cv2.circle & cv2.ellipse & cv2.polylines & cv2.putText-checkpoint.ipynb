```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## line

```python
img_blank = np.zeros(shape=[512, 512, 3]).astype(np.uint8)
cv2.line(img_blank, (0, 0), (511, 511), color=[255, 0, 0], thickness=5)

fig, ax = plt.subplots(1, figsize=[16, 10])
ax.imshow(img_blank)
ax.set_axis_off()
plt.show()
```

## rectangle

```python
# pt1:左上角坐标[xmin, ymin] ; pt2:右下角坐标[xmax, ymax]
cv2.rectangle(img_blank, pt1=(156, 156), pt2=(356, 356), color=[0, 255, 0], thickness=5)

fig, ax = plt.subplots(1, figsize=[16, 10])
ax.imshow(img_blank)
ax.set_axis_off()
plt.show()
```

## circle

```python
# center:圆心； radius:半径
cv2.circle(img_blank, center=(255, 255), radius=100, color=[0, 0, 255], thickness=5)

fig, ax = plt.subplots(1, figsize=[16, 10])
ax.imshow(img_blank)
ax.set_axis_off()
plt.show()
```

## ellipse

```python
# center:椭圆的圆心坐标； 
# axes：一个包含两个元素的元组，分别表示椭圆的长轴和短轴长度
# angle:椭圆沿逆时针方向旋转的角度
# startAngle：表示椭圆圆弧沿顺时针方向的起始角度
# endAngle:表示椭圆圆弧沿顺时针方向的结束角度
# 如果startAngle不等于0且endAngle不等于360，那么最终在图形上只显示部分圆弧，startAngle和endAngle两个参数允许用户只画出部分圆弧

cv2.ellipse(img_blank, center=(255, 255), axes=(100,50), angle=0, startAngle=0, endAngle=360, color=[0, 255, 255], thickness=5)

fig, ax = plt.subplots(1, figsize=[16, 10])
ax.imshow(img_blank)
ax.set_axis_off()
plt.show()
```

## ploylines

```python
# pts:多边形的各定点（dtype=int32），每个顶点坐标的形式为[x, y]
# opencv将传入的各店依次用直线连接起来
pts = np.array([[255, 155], [155, 255], [255, 355], [355, 255]]).astype(np.int32)
pts = pts.reshape([-1, 1, 2])
cv2.polylines(img_blank, pts=[pts], isClosed=True, color=[255, 0, 255], thickness=5)

fig, ax = plt.subplots(1, figsize=[16, 10])
ax.imshow(img_blank)
ax.set_axis_off()
plt.show()
```

## putText

```python
# text:显示的文本
# org文本框左下角坐标（只接受元素为int的元组）
# fontFace：字体类型
# fontScale:字体大小（float）
# thickness：int，值为-1时表示填充颜色
font = cv2.FONT_HERSHEY_SIMPLEX
text = 'Opencv'
cv2.putText(img_blank, text=text, org=(255, 255), fontFace=font, fontScale=1, color=[255, 0, 255], thickness=1)

fig, ax = plt.subplots(1, figsize=[16, 10])
ax.imshow(img_blank)
ax.set_axis_off()
plt.show()
```
