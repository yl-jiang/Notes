## 使用BRIEF计算特征描述符

推荐使用SURF特征检测器配合BRIEF

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./imgdata/cpp1.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 创建star特征计算子
star = cv2.xfeatures2d.StarDetector_create()

# 创建BRIEF计算子
brief = cv2.xfeatures2d.BriefDescriptorExtractor_create()

key_points = star.detect(img_gray, None)

# 展示关键点
img_copy = img_rgb.copy()
img_copy = cv2.drawKeypoints(img_copy, key_points, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.figure(figsize=[8, 8])
plt.imshow(img_copy)
plt.show()
```

```python
key_points, des = brief.compute(img_gray, key_points)
# 展示关键点
img_copy = img_rgb.copy()
img_copy = cv2.drawKeypoints(img_copy, key_points, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.figure(figsize=[8, 8])
plt.imshow(img_copy)
plt.show()
```

```python
print(brief.descriptorSize())
```
