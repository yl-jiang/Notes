**FastFeatureDetector算法比其他角点检测算法快，但是在噪声很高的情况下不够稳定。**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

`cv2.FastFeatureDetector_create([, threshold[, nonmaxSuppression[, type]]])`

+ threshold:
+ nonmaxSuppression:是否使用非极大值抑制解决检测到的特征点相连的问题【默认使用】

```python
img = cv2.imread('./imgdata/black_white_card.jpg')

fast = cv2.FastFeatureDetector_create(threshold=25)

key_points = fast.detect(img, None)

for point in key_points:
    xy = (int(point.pt[0]), int(point.pt[1]))
    cv2.circle(img, xy, 10, 255, -1)
    
# img = cv2.drawKeypoints(img, key_points, None, color=(255, 0, 0))

plt.figure(figsize=[16, 10])
plt.imshow(img)
plt.show()
```

```python
img = cv2.imread('./imgdata/black_white_card.jpg')

fast = cv2.FastFeatureDetector_create(threshold=25)
fast.setNonmaxSuppression(0)
key_points = fast.detect(img, None)

# 关闭默认开启的非极大值抑制
img = cv2.drawKeypoints(img, key_points, None, color=(255, 0, 0))

plt.figure(figsize=[16, 10])
plt.imshow(img)
plt.show()
```

## 使用SIFT算法寻找角点

```python
img = cv2.imread('./imgdata/black_white_card.jpg')

sift = cv2.xfeatures2d.SIFT_create()
key_points = sift.detect(img, None)

for point in key_points:
    if point.size > 50:
        xy = (int(point.pt[0]), int(point.pt[1]))
        cv2.circle(img, xy, 10, 255, -1)
    
# img = cv2.drawKeypoints(img, key_points, None, color=(255, 0, 0))

plt.figure(figsize=[16, 10])
plt.imshow(img)
plt.show()
```
