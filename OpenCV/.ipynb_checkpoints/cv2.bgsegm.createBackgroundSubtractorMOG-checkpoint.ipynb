```python
import cv2
import numpy as np
```

```python
video = cv2.VideoCapture('./imgdata/car.flv')
video.set(cv2.CAP_PROP_POS_FRAMES, 300)
fgbg = cv2.bgsegm.createBackgroundSubtractorMOG()

mark = 1
for _ in range(165):
    ret, frame = video.read()
    fgmask = fgbg.apply(frame)
    cv2.imwrite(f'./background_substract/img_{mark}.jpg', fgmask)
    mark += 1
```

```python
from pathlib import Path
import matplotlib.pyplot as plt
from PIL import Image

result = Path('./background_substract/')

for p in result.iterdir():
    if p.is_file():
        img = np.asarray(Image.open(str(p)))
        plt.imshow(img, cmap='gray')
        plt.show()
```
