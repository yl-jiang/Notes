# 分水岭算法图像分割

任何一副灰度图像都可以被看成拓扑平面，灰度值高的区域可以被看成是山峰，灰度值低的地方可以被看成是山谷。我们向每一个山谷中灌不同颜色的水，随着水位升高，不同山谷的水就会相遇汇合，为了防止不同山谷的水汇合，我们需要在汇合处建起大坝，然后再不停地灌水，不停地构建堤坝知道所有山峰都被水淹没。我们构建的堤坝就是对图像的分割。

这种方法会因为噪声或者图像中其他不规律的因素造成得到过度分割的结果，为了减少这种影响，opencv采用了基于mask的分水岭算法，在这种算法中我们要设置哪些山谷点会汇合，哪些不会。这是一种交互式的图像分割。我们要做的就是给我们一致的对象打上不同的标签。如果某个区域肯定不是对象卫视背景，就是用另外一种颜色标签标记，而剩下的不能确定的前景还是背景的区域就用0标记。标记完后，使用分水岭算法，每一次灌水，我们的标签就会被更新，当不同颜色的标签相遇时就构建堤坝，直至将所有山峰淹没，最后我们得到的边界对象的值为-1.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
def show(img, cmap=None):
    if cmap is None:
        cmap = 'gray'
    plt.figure(figsize=[8, 8])
    plt.imshow(img, cmap=cmap)
    plt.show()
```

```python
matplotlib.cm.get_cmap
```

## 获取灰度图像

```python
img = cv2.imread('imgdata/coins.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

show(img_rgb, 'plasma')
show(img_gray)
```

## 二值化灰度图像

```python
ret, thresh = cv2.threshold(img_gray, 200, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
fig = np.hstack((img_gray, thresh))

show(thresh, 'gray')
```

## 形态学操作，消除图像中的噪点

```python
# 闭运算填充物体内细小空洞、连接邻近物体
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
fg = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel ,iterations=2)

show(fg, 'gray')
```

## 膨胀操作，进一步扩大前景区域，并确定背景区域

白色部分是前景（主体，我们需要分割出来的区域），黑色部分是背景区域。在这一步中我们只关心黑色区域，因为它是我们很确定的背景区域。

```python
sure_bg = cv2.dilate(fg, kernel, iterations=3)
show(sure_bg, 'gray')
```

```python
np.unique(sure_bg)
```

## 使用距离变换获取确定的前景色

+ 根据distanceTransform获取距离背景最小距离的结果
+ 根据distanceTransform操作的结果，设置一个阈值，使用threshold决定哪些区域是前景，这样得到正确结果的概率很高

在这一步中我们只关心白色区域，因为这是我们很确定的前景区域。其实前景区域是我们的种子，我们将从这里进行灌水，向四周涨水。

```python
dist = cv2.distanceTransform(sure_bg, cv2.DIST_L2, 5)
ret, sure_fg = cv2.threshold(dist, dist.max()*0.6, 255, cv2.THRESH_BINARY)

# 因为distanceTransform函数返回的结果是浮点数，为了显示，现将其归一化到0-1
normal_dist = cv2.normalize(dist, 0, 1.0, cv2.NORM_MINMAX)
show(normal_dist * 100)
show(sure_fg)
```

距离变换的处理图像通常都是二值图像，而二值图像其实就是把图像分为两部分，即背景和物体两部分，物体通常又称为前景目标！

通常我们把前景目标的灰度值设为255，即白色

背景的灰度值设为0，即黑色。

所以定义中的非零像素点即为前景目标，零像素点即为背景。

所以图像中前景目标中的像素点距离背景越远，那么距离就越大，如果我们用这个距离值替换像素值，那么新生成的图像中这个点越亮。

![distanceTransform](./imgdata/distanceTransform.jpg)

再通过设定合理的阈值对距离变换后的图像进行二值化处理，则可得到去除手指的图像(如下图“bidist”窗口图像所示)，手掌重心即为该图像的几何中心。

![distanceTransform2](./imgdata/distanceTransform2.jpg)

`distanceTransform`函数主要用于计算非零像素到最近零像素点的最短距离。一般用于求解图像的骨骼

`cv2.distanceTransform(src, distanceType, maskSize, dst=None, dstType=None):`

+ src：输入的图像，一般为二值图像
+ distanceType：所用的求解距离的类型，有CV_DIST_L1, CV_DIST_L2 , or CV_DIST_C
+ mask_size：距离变换掩模的大小，可以是 3 或 5. 对 CV_DIST_L1 或 CV_DIST_C 的情况，参数值被强制设定为 3, 因为 3×3 mask 给出 5×5 mask 一样的结果，而且速度还更快。

`distanceTransform`返回的图像数据是浮点数值，要想在浮点数表示的颜色空间中，数值范围必须是0-1.0，所以要将其中的数值进行归一化处理

## 获取未知区域

```python
fig, axes = plt.subplots(1, 2, figsize=[16, 8])
axes[0].set_title('forntground', fontsize=15)
axes[0].imshow(sure_fg, cmap='gray')
axes[1].set_title('background', fontsize=15)
axes[1].imshow(sure_bg, cmap='gray')
plt.show()
```

上面frontground中的白色区域是我们非常确定的前景区域，background中的黑色区域是我们非常确定的背景区域。frontground和background颜色不一致的区域就是未知区域。

```python
sure_fg.dtype, sure_bg.dtype
```

```python
sure_fg_uint8 = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg, sure_fg_uint8)
show(unknown)
```

上图中的白色区域就是unknown区域。

## 创建标签

对我们已经确定分类的区域（无论是前景还是背景）使用不同的正整数标记，对我们不确定的区域（unknown区域）使用 0 标记。

我们可以使用函数 `cv2.connectedComponents()`来做这件事。利用connectedComponents求图中的连通图

`cv2.connectedComponents(image, labels=None, connectivity=None, ltype=None)`

+ 参数image是需要进行连通域处理的二值图像，其他的这里用不到

connectedComponents函数会将其认为属于同一个联通的区域像素设置为同一个像素值，不同的连通区域的值不同。

```python
ret, markers = cv2.connectedComponents(sure_fg_uint8)

show(markers)
```

ret是连通域处理的边缘条数，是上面提到的确定区域（除去背景外的其他确定区域：也就是前景），就是种子数，我们会从种子开始向外涨水。

markers是我们创建的一个标签（一个与原图像大小相同，数据类型为 in32 的数组），其中包含有我们原图像的确认区域的数据（前景区域）

但上述由connectedComponents得到的markers并不是我们所希望的，因为我们期望只有unknown区域的像素值为0。

### 修正markers

```python
markers += 1
markers[unknown == 255] = 0

show(markers)
```

## 水漫金山

OpenCV的分水岭算法`cv::wathershed`，需要输入一个标记图像，图像的像素值为32位有符号正数（`CV_32S`类型），每个非零像素代表一个标签。它的原理是对图像中部分像素做标记，表明它的所属区域是已知的。分水岭算法可以根据这个初始标签确定其他像素所属的区域。

本质上讲，基于标记点的改进算法是利用先验知识来帮助分割的一种方法。因此，改进算法的关键在于如何获得准确的标记图像，即如何将前景物体与背景准确的标记出来。

```python
segment_markers = cv2.watershed(img_rgb, markers)
img_rgb[segment_markers == -1] = [255, 0, 0]

show(img_rgb, 'plasma')
```

```python
tmp = np.full_like(img_rgb, fill_value=255, dtype=np.uint8)
tmp[segment_markers == -1] = [125, 0, 125]
show(tmp, 'plasma')
```
