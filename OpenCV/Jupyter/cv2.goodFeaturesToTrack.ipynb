`cv2.goodFeaturesToTrack(image, maxCorners, qualityLevel, minDistance[, corners[, mask[, blockSize[, useHarrisDetector[, k]]]]])`

+ image: 灰度图像
+ maxCorners: 返回角点的个数
+ qualityLevel: 角点的质量，介于0-1之间，低于该值的角点将被忽略
+ minDistance: 角点之间最短的欧氏距离

return
+ 返回的是各角点的坐标

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./imgdata/black_white_card.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

corners = cv2.goodFeaturesToTrack(img_gray, 200, 0.01, 10)
corners = np.int0(corners)

fig = img.copy()

for i in corners:
    x, y = i.ravel()
    cv2.circle(fig, (x, y), 15, 255, -1)

plt.figure(figsize=[16, 10])
plt.imshow(fig)
plt.show()
```
