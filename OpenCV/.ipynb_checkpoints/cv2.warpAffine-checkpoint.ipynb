```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
%matplotlib inline
```

```python
img_path_1 = './imgdata/img.jpg'
img_1 = cv2.imread(img_path_1)
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)

fig, ax = plt.subplots(1, figsize=[16, 10])
# ax.set_axis_off()
ax.imshow(img_1)
plt.show()
```

$$如果你要沿(x, y)方向移动，对应的移动距离为(t_x, t_y)，可以用下面的方式构建移动矩阵：$$
$$\begin{bmatrix}1 & 0 & t_x\\\\0 & 1 & t_y\end{bmatrix}$$
矩阵数据类型为np.float32

#### cv2.warpAffine(src, M, dsize[, dst[, flags[, borderMode[, borderValue]]]]) 
+ src - 输入图像。
+ M - 变换矩阵。
+ dsize - 输出图像的大小[width, heigth]。
+ flags - 插值方法的组合（int 类型！）
+ borderMode - 边界像素模式（int 类型！）
+ borderValue - （重点！）边界填充值; 默认情况下，它为0。

```python
# 使图像沿x轴方向向右移动100像素，沿y轴方向向下移动50像素
translate_matrix = np.array([[1, 0, 100], [0, 1, 50]]).astype(np.float32)
h, w, c = img_1.shape
img_translate = cv2.warpAffine(src=img_1, M=translate_matrix, dsize=(w, h), borderValue=[255, 125, 0])

# show
fig, ax = plt.subplots(1, figsize=[16, 10])
# ax.set_axis_off()
ax.imshow(img_translate)
plt.show()
```

```python
# 使图像沿x轴方向向左移动100像素，沿y轴方向向上移动50像素
translate_matrix = np.array([[1, 0, -100], [0, 1, -50]]).astype(np.float32)
h, w, c = img_1.shape
img_translate = cv2.warpAffine(src=img_1, M=translate_matrix, dsize=(w, h), borderValue=[255, 125, 0])

# show
fig, ax = plt.subplots(1, figsize=[16, 10])
# ax.set_axis_off()
ax.imshow(img_translate)
plt.show()
```
