# 模板匹配

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

模板匹配是用来在一副大图中搜寻查找模板图像位置的方法。opencv为我们提供了函数`cv2.matchTemplate()`,和2D卷积一样，他也是用模板图像再输入图像上滑动，并在每一个位置对模板图像和其对应的输入图像的子区域进行比较。同时，opencv在这个函数中提供了几种不同的比较方法，返回的结果是一个灰度图像，每一个像素值表示了此区域与模板的匹配程度。

如果输入图像的大小是WxH，模板的大小是wxh，输出的结果的大小就是(W-w+1)x(H-h+1)。当我们得到`cv2.matchTemplate()`的输出结果，就可以使用`cv2.minMaxLoc()`来找到其中的最小值和最大值了。`cv2.minMaxLoc()`返回的第一个元素为矩形左上角的点（位置），(w, h)为模板矩形的宽和高。这个矩形就是找到的模板区域了。

```python
img = cv2.imread('./imgdata/messi.jpg')
cv2.cvtColor(src=img, code=cv2.COLOR_BGR2RGB, dst=img)
plt.figure(figsize=[16, 10])
plt.imshow(img)
plt.show()
```

```python
template = img[250:390, 640:750, :]
plt.imshow(template)
plt.show()
```

```python
methods = ['cv2.TM_CCOEFF', 
           'cv2.TM_CCOEFF_NORMED', 
           'cv2.TM_CCORR', 
           'cv2.TM_CCORR_NORMED', 
           'cv2.TM_SQDIFF', 
           'cv2.TM_SQDIFF_NORMED']

w, h = template.shape[:2]

for meth in methods:
    img_copy = img.copy()
    method = eval(meth)
    res = cv2.matchTemplate(image=img_copy, templ=template, method=method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
    # 特别处理两种特殊的比较方法
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
    bottom_right = (top_left[0] + w, top_left[1] + h)
    cv2.rectangle(img_copy, top_left, bottom_right, (0, 255, 255), 3)
    fig, axes = plt.subplots(1, 2, figsize=[20, 5])
    fig.suptitle(meth, fontsize=16)
    axes[0].imshow(res, cmap='gray')
    axes[1].imshow(img_copy)
    plt.show()
```
