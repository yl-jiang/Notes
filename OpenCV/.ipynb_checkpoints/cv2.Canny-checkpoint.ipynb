```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('./anime.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots(figsize=[8, 10])
ax.imshow(img)
plt.show()
```

#### cv2.Canny(image, threshold1, threshold2[, edges[, apertureSize[, L2gradient]]])

+ threshold1minVal
+ threshold2maxVal
+ apertureSize: Sobel
+ L2gradient:L2L1

```python
img_canny = cv2.Canny(image=img, threshold1=50, threshold2=200, apertureSize=5, L2gradient=True)

fig, ax = plt.subplots(1, 2, figsize=[12, 8])
ax[0].imshow(img)
ax[0].set_title('Original')
ax[1].imshow(img_canny, cmap='gray')
ax[1].set_title('Canny')
plt.show()
```
