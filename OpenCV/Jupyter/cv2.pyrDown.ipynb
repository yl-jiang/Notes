```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('./anime.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots(figsize=[8, 10])
ax.imshow(img)
plt.show()
```

+ src:

```python
fig, ax = plt.subplots(figsize=[16, 20])
ax.imshow(img)
plt.show()
for i in range(4):
    img_lr = cv2.pyrDown(img)
    img = img_lr
    fig, ax = plt.subplots(figsize=[16//((i+1)*2), 20//((i+1)*2)])
    ax.imshow(img_lr)
    plt.show()
```
