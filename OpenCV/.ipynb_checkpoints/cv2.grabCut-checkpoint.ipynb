## 使用GrabCut算法进行交互式前景提取

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
def show(img, cmap=None):
    if cmap is None:
        cmap = 'gray'
    plt.figure(figsize=[8, 8])
    plt.imshow(img, cmap=cmap)
    plt.show()
```

```python
img = cv2.imread('./imgdata/messi.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
show(img_rgb, 'plasma')
```

```python
img.shape
```

`cv2.grabCut()`

+ img: 输入图像
+ mask: 用来确定哪些区域是前景，背景，可能是前景，背景；可设置为：cv2.GC_BGD, cv2.GC_FGD, cv2.GC_PR_BGD, cv2.GC_PR_FGD,或者直接输入0,1,2,3
+ rect:包含前景的矩形框（格式为[x, y, w, h]）
+ bgdModel, fgdModel:算法内部使用的数组，你只需要创建两个大小为(1, 65)，数据类型为np.float64的数组
+ iterCount: 算法的迭代次数
+ mode：可设置为cv2.GC_INIT_WITH_RECT或者cv2.GC_INIT_WITH_MASK,也可以联合使用，这是用来确定我们进行修改的方式，矩形模式或者mask模式

### 使用矩形模式

```python
import matplotlib.patches as patches

mask = np.zeros(img.shape[:2], np.uint8)
bgdModel = np.zeros((1, 65), np.float64)
fgdModel = np.zeros((1, 65), np.float64)

xyxy = [100, 100, 1400, 995] #  [xmin, ymin, xmax, ymax]
fig, ax = plt.subplots(figsize=[8, 8])
rect = patches.Rectangle(xy=(xyxy[0], xyxy[1]), 
                         width=xyxy[2]-xyxy[0], 
                         height=xyxy[-1]-xyxy[1], 
                         fill=False,
                         color='r')
ax.imshow(img_rgb)
ax.add_patch(rect)
plt.show()
```

grabCut算法会inplace修改传入的mask参数，返回的新的mask中所有像素被分为4类：背景，前景，可能是背景和可能是前景，分别取值0,1,2,3

```python
xywh = ((xyxy[2]-xyxy[0])/2+1, (xyxy[3]-xyxy[1])/2+1, xyxy[2]-xyxy[0], xyxy[3]-xyxy[1])
xywh = tuple([int(_) for _ in xywh])

cv2.grabCut(img_rgb, mask, xywh, bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_RECT)
```

将更新后mask中代表背景和可能是背景的像素值设为0，其他像素值设为1.用于显示前景和可能是前景的原图像像素。

```python
mask_update = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')

img_seg = img_rgb * mask_update[:, :, None]
show(img_seg, 'plasma')
plt.show()
```
