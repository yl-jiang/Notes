## 直方图均衡化

使用直方图均衡化改善图片的对比。

## 使用numpy实现直方图均衡化

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./imgdata/hist.jpg', flags = 0)
```

```python
plt.imshow(img, cmap='gray')
plt.show()
```

```python
def histogram(img):
    hist, bins = np.histogram(img.ravel(), bins=256, range=[0,256])
    # 累积分布图
    cdf = hist.cumsum()
    cdf_normalized = cdf * hist.max() / cdf.max()
    plt.plot(cdf_normalized, c= 'b')
    plt.hist(img.ravel(), bins=256, range=[0,256], color='r')
    plt.xlim([0,256])
    plt.legend(('cdf', 'histogram'), loc='upper left')
    plt.show()
```

```python
histogram(img)
```

直方图的大部分在灰度值比较高的部分，而且分布比较集中，而一副高质量的图像的灰度值分布是比较广泛的。我们希望改变图片像素的亮度分布，使得改变后的图像的直方图能够尽可能的覆盖整个x轴。

```python
cdf_mask_1 = np.ma.masked_where(cdf == 0, cdf)
cdf_mask_2 = np.ma.masked_equal(cdf, 0)
np.allclose(cdf_mask_1, cdf_mask_2)
```

```python
cdf_mask_1
```

可以看到，img上没有小于像素值为123的像素。

```python
cdf_mask_1.min()
```

```python
cdf_mask_1 = (cdf_mask_1 - cdf_mask_1.min()) * 255 / (cdf_mask_1.max() - cdf_mask_1.min())
cdf_filled = np.ma.filled(cdf_mask_1, 0).astype('uint8')
cdf_filled
```

```python
len(cdf_filled[cdf_filled == 0]), len(np.unique(cdf_filled))
```

直方图均衡化后的图像只有68种不同的像素值。

```python
# 将img上的每个像素值都当做索引，即cdf_filled是一张从输入到输出的像素值映射表

# 将img上像素值小于123的像素值置为0(img上不存在这样的像素)，
# 等于124的像素置为1（img上只有一个这样的像素），
# 等于125的像素置为2（img上只有一个这样的像素），
# 等于126的像素置为3（img上只有一个这样的像素）， 
# 等于127的像素置为4（img上只有一个这样的像素）， 
# 等于128的像素置为5（img上只有一个这样的像素）， 
# 等于129的像素置为7（img上只有两个这样的像素）， 
# ...

img2 = cdf_filled[img]
img2.shape
```

```python
plt.imshow(img2, cmap='gray')
plt.show()
```

```python
histogram(img2)
```

**直方图均衡化是经常用来使所有的图片具有相同的亮度条件的参考工具**

## 使用opencv的直方图均衡化方法

`cv2.equalizeHist(src[, dst])`

输入是一张灰度图像，输出是直方图均衡化之后的图像。

```python
img = cv2.imread('./imgdata/hist.jpg', 0)
equ = cv2.equalizeHist(img)
res = np.hstack([img, equ])
plt.imshow(res, cmap='gray')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error as mse

mse(equ, img2)
```

```python
histogram(equ)
```

上述直方图均衡化方法，适用于图像直方图分布比较集中，且x轴覆盖范围比较窄的情形。

## 自适应直方图均衡化

```python
img = cv2.imread('imgdata/tsukuba.jpg', flags=0)
plt.imshow(img, cmap='gray')
plt.show()
```

```python
histogram(img)
```

上幅图像的直方图分布比较均匀，此时使用普通的直方图均衡化：

```python
equ = cv2.equalizeHist(img)
res = np.hstack([img, equ])
plt.figure(figsize=[10, 5])
plt.imshow(res, 'gray')
plt.show()
```

直方图均衡化后，背景的对比度得到了加强，但是雕像的细节损失严重。

```python
histogram(equ)
```

为了解决这个问题，我们需要使自适应的直方图均衡化。这种情况下，整幅图像会被分成很多小块，这些小块被称为”tiles“（opencv中tiles的大小默认为8x8），然后再对每一个小块分别进行直方图均衡化（跟之前的类似）。所以在每一个区域中，直方图会集中在某一个小的区域中（除非有噪声干扰）。如果有噪声的话，噪声会被放大。为了避免这种情况的出现，要使用对比度限制。对于每小块来说，如果**直方图的bin**超过对比度上限的话，酒吧其中的像素点均匀分散到其它bins中，然后再进行直方图均衡化。最后，为了去除每一小块之间”人造“边界，再使用双线性插值，对小块进行缝合。

`createCLAHE([, clipLimit[, tileGridSize]])`

brief Creates implementation for cv::CLAHE .

+ clipLimit Threshold for contrast limiting.[对比度限制阈值]
+ tileGridSize Size of grid for histogram equalization. Input image will be divided into

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl = clahe.apply(img)
plt.imshow(cl, 'gray')
plt.show()
```

```python
histogram(cl)
```
