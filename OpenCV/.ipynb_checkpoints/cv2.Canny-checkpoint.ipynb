```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('./anime.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots(figsize=[8, 10])
ax.imshow(img)
plt.show()
```

### Canny边缘检测是一种非常流行的边缘检测算法，具体步骤如下：
1. 去噪
    边缘检测很容易受到噪声的影响，第一步先使用5x5的高斯滤波器去除噪声
2. 计算图像梯度
    对平滑后的图像使用Sobel算子计算水平和垂直方向的一阶导数，根据得到的两幅梯度图找到边界的梯度和方向（边界的梯度方向总是与边缘垂直，其中梯度的方向被归为4类：垂直，水平，以及两对角线）
3. 非极大值抑制
    获得梯度的方向和大小后，对整幅图进行扫描，去除那些非边界上的点。【具体做法：对每一个像素进行检查，看这个点的梯度是不是周围具有相同方向的点中最大的】
4. 滞后阈值
    确定哪些边界才是真正的边界。这时我们需要设置两个阈值：minVal和maxVal。当图像灰度梯度高于maxVal时被认为是真边界，低于minVal的边界会被抛弃。如果介于两者之间，那么考察这个点是不是与某个确定为真正边界的像素相连，如果是则该点被认为是边界点，否则就抛弃。在这一步中一些小的噪声会被去除，这一步的前提假设是我们认为边界都是一些长的线段。【选择合适的minVal和maxVal是非常重要的】
    
#### cv2.Canny(image, threshold1, threshold2[, edges[, apertureSize[, L2gradient]]])
+ image:输入图像
+ threshold1：minVal
+ threshold2：maxVal
+ apertureSize: Sobel算子的大小
+ L2gradient:一个布尔值，如果为真，则使用更精确的L2范数进行计算（即两个方向的倒数的平方和再开放），否则使用L1范数（直接将两个方向导数的绝对值相加）

```python
img_canny = cv2.Canny(image=img, threshold1=50, threshold2=200, apertureSize=5, L2gradient=True)

fig, ax = plt.subplots(1, 2, figsize=[12, 8])
ax[0].imshow(img)
ax[0].set_title('Original')
ax[1].imshow(img_canny, cmap='gray')
ax[1].set_title('Canny')
plt.show()
```
