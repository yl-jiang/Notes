```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_path = '/home/dk/jyl/Data/COCO2017/val2017/000000001296.jpg'

img = cv2.imread(img_path)
print(img.shape)
```

```python
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img = cv2.rectangle(img=img_rgb,
              pt1=(20, 20),
              pt2=(100,100),
              color=(255, 255, 0),
              thickness=2)
```

```python
img = cv2.putText(img=img, 
                  text='test', 
                  org=(100, 100), 
                  fontFace=cv2.FONT_ITALIC, 
                  fontScale=1.5,
                  color=(0, 255, 0))
```

```python
plt.figure(figsize=[16, 9])
plt.imshow(img)
plt.show()
```
