一维直方图中我们只考虑了图像的灰度这一个特征，在2D直方图中我们要考虑图像的两个特征。对于彩色图像的2D直方图通常考虑颜色（Hue）和饱和度（Saturation）。

为了得到颜色和饱和度的数值，首先，要将图片从BGR转换到HSV（而1D直方图需要从BGR转换为gray）。

计算2D直方图，仍使用`cv2.calcHist()`函数，但是函数的参数要做如下设置：

1. `channels = [0, 1]`，因为我们需要同时处理H和S通道
2. `histSize = [180, 256]`，因为H通道的取值范围为0-180，S通道的取值范围在0-256
3. `ranges = [0, 180, 0, 256]`，因为H通道的取值范围为0-180，S通道的取值范围在0-256

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./imgdata/img.jpg')
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb, cmap='hsv')
plt.show()

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv, cmap='hsv')
plt.show()
```

## 使用opencv计算2D直方图

```python
img = cv2.imread('./imgdata/img.jpg')
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
hist = cv2.calcHist([hsv], channels=[0, 1], mask=None, histSize=[180, 256], ranges=[0, 180, 0, 256])
```

```python
hist.shape
```

## 使用numpy计算2D直方图

```python
img = cv2.imread('./imgdata/img.jpg')
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
hist, xbins, ybins = np.histogram2d(hsv[..., 0].ravel(), hsv[..., 1].ravel(), bins=[180, 256], range=[[0,180],[0,256]])
```

```python
hist.shape
```

## show 2D直方图

```python
img = cv2.imread('./imgdata/img.jpg')
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
hist = cv2.calcHist([hsv], channels=[0, 1], mask=None, histSize=[180, 256], ranges=[0, 180, 0, 256])

plt.imshow(hist, interpolation='nearest', cmap='hsv')
plt.show()
```
