```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
%matplotlib inline
```

```python
img_path_1 = './img.jpg'
img_1 = cv2.imread(img_path_1)
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)

fig, ax = plt.subplots(1, figsize=[16, 10])
# ax.set_axis_off()
ax.imshow(img_1)
plt.show()
```

### opencv的HSV格式中，H（色彩、色度）的取值范围为[0, 179]，S（饱和度）的取值范围为[0, 255],V（亮度）的取值范围为[0, 255]，但是不同的软件HSV的取值范围可能不同

HSV色彩空间要比RGB更容易表示一个特定颜色

#### cv2.inRange(hsv, lowerb, upperb)

+ 第一个参数：hsv指的是原图

+ 第二个参数：lower_red指的是图像中低于这个lower_red的值，返回的mask对应的值变为0

+ 第三个参数：upper_red指的是图像中高于这个upper_red的值，返回的mask对应的值变为0

+ 而在lower_red～upper_red之间的值变成255

```python
hsv = cv2.cvtColor(img_1, cv2.COLOR_RGB2HSV)
# 设置目标颜色阈值（当某个像素的颜色落在某一范围内时就认为它为蓝色）
low_blue = np.array([0, 43, 46])
hig_blue = np.array([10, 255, 255])
# 根据颜色阈值构建mask
mask = cv2.inRange(hsv, low_blue, hig_blue)
fig, ax = plt.subplots(1, figsize=[16, 10])
# ax.set_axis_off()
ax.imshow(mask, cmap='gray')
plt.show()
```

```python
# 对目标图像根据mask进行位运算,只显示原图中指定颜色范围内的像素
res = cv2.bitwise_and(src1=img_1, src2=img_1, mask=mask)
fig, ax = plt.subplots(1, figsize=[16, 10])
# ax.set_axis_off()
ax.imshow(res)
plt.show()
```
