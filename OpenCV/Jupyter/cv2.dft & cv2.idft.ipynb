# 图像变换

在图像问题当中，频域是指图像的灰度变化，也就是灰度图像的梯度值，这个和轮廓的原理差不多，灰度值变化剧烈的叫做高频分量，例如边界和噪声。灰度值变化缓慢的称谓低频分量.

对图像进行X方向和Y方向的傅里叶变换，会得到图像的频域表示图。

## 傅里叶变换

对于一个正弦信号，如果它的幅度变化非常快，我们可以说他是高频信号，如果变化非常慢，我们称之为低频信号。把这种想法应用到图像中，图像哪里的幅度变化非常大呢？边界点或者噪声点。所以我们说边界和噪声是图像中的高频分量（这里的高频是指变化非常快，为非出现次数多）。如果没有如此大的幅度变化我们称之为低频分量。

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## numpy中的傅里叶变换

2D傅里叶变换F(x,y)的F(0,0)位置在图像的左上角，F(0,0)表示的是图像灰度的均值，由于该位置不利于观察，通常将该点移到图像的中心区域进行观察。

```python
img = cv2.imread('./imgdata/messi.jpg', 0)
# np.fft.fft2函数可以对信号进行频率转换，输出的结果是一个复数矩阵，叫做图像的频谱图
# 矩阵中的每个值对应F(u,v)，由于是复数，所以一般在数学计算上都用极坐标表示。
f = np.fft.fft2(img)
# 对np.fft.fft2得到的频率进行向下和向右平移h/2,w/2
fshift = np.fft.fftshift(f)
# 构建振幅图
magnitude_spectrum = 20 * np.log(np.abs(fshift))

fig, axes = plt.subplots(1, 2, figsize=[16, 10])
axes[0].imshow(img, cmap='gray')
axes[1].imshow(magnitude_spectrum, cmap='gray')
plt.show()
```

```python
fshift.shape, fshift.dtype
```

上面输出的振幅图的中心部分更白，说明低频分量多。下面就得到的频率的基础上进行频域变换，例如使用高通滤波和重建图像（DFT的逆变换）。高通滤波的意思就是增强边缘，非边缘部分被过滤，同理，低通滤波就是将图像边的平滑

```python
rows, cols = img.shape
crow, ccol = rows // 2, cols // 2
# 使用60x60的窗口对图像进行掩模操作，从而去除低频分量
fshift[crow-30:crow+30, ccol-30:ccol+30] = 0
# np.fft.ifftshift逆平移变换（'i' means inverse）
# 使用FFT逆变换，此时结果仍然是复数
f_ishift = np.fft.ifftshift(fshift)
# np.fft.ifft2函数为FFT的逆变换（'i' means inverse）

img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)

fig, axes = plt.subplots(1, 2, figsize=[16, 20])
axes[0].imshow(img, cmap='gray')
axes[1].imshow(img_back, cmap='gray')
plt.show()
```

```python
f_ishift.shape, f_ishift.dtype
```

上面例子可以看到高通滤波和边缘检测差不多。

## 使用opencv做傅里叶变换

`cv2.dft(src[, dst[, flags[, nonzeroRows]]])`

+ src: 输入图像，灰度图像，dtype为float32

```python
img = cv2.imread('./imgdata/messi.jpg', 0)
dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[..., 0], dft_shift[..., 1]))

fig, axes = plt.subplots(1, 2, figsize=[16, 10])
axes[0].imshow(img, cmap='gray')
axes[1].imshow(magnitude_spectrum, cmap='gray')
plt.show()
```

接下来做低通滤波，与高通滤波相对应，低通滤波会对图像进行模糊处理。

```python
rows, cols = img.shape
crow, ccol = rows // 2, cols // 2

# 构建一个mask，低频对应的区域设为1，高频区域设为0
mask = np.zeros((rows, cols, 2), np.uint8)
mask[crow-30:crow+30, ccol-30:ccol+30] = 1

fshift = dft_shift * mask
# 将频谱图还原（中心点回到左上角）
f_ishift = np.fft.ifftshift(fshift)
# 对频谱图进行逆向FFT变换
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[..., 0], img_back[..., 1])

fig, axes = plt.subplots(1, 2, figsize=[16, 10])
axes[0].imshow(img, cmap='gray')
axes[1].imshow(img_back, cmap='gray')
plt.show()
```

## DFT的性能优化

当数组的大小是2,3,5的倍数时DFT的效率较高，如果想提高代码的运行效率，可以修改输入图像的大小（补0）.opencv需要手动补0，numpy自动。

怎样确定最佳的大小？opencv提供的函数`cv2.getOptimalDFTSize()`可以帮助我们。

```python
img = cv2.imread('./imgdata/messi.jpg', 0)
img = cv2.resize(src=img, dsize=(548, 342))
rows, cols = img.shape
print(rows, cols)
```

```python
optimal_rows = cv2.getOptimalDFTSize(rows)
optimal_cols = cv2.getOptimalDFTSize(cols)
print(optimal_rows, optimal_cols)
```

```python
img_pad = np.zeros(shape=(optimal_rows, optimal_cols))
img_pad[:rows, :cols] = img
```

### numpy

```python
%timeit fft1 = np.fft.fft2(img)
```

```python
%timeit fft2 = np.fft.fft2(img, [optimal_rows, optimal_cols])
```

### opencv

```python
%timeit dft1 = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
```

```python
%timeit dft1 = cv2.dft(np.float32(img_pad), flags=cv2.DFT_COMPLEX_OUTPUT)
```

无论是padding之前还是之后，opencv的速度都要比numpy快很多。

## 为什么拉普拉斯算子是高通滤波器

```python
mean_filter = np.ones(shape=(3,3))

# guassian filter
x = cv2.getGaussianKernel(5, 10)
gaussian = x * x.T

# different edge detecting filters

# scharr in x-dimension
scharr = np.array([[-3, 0, 3], [-10, 0, 10], [-3, 0, 3]])

# sobel in x-dimension
sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
# sobel in y-dimension
sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

laplacian = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])

filters = [mean_filter, gaussian, laplacian, sobel_x, sobel_y, scharr]
filter_name = ['mean_filter', 'gaussian', 'laplacian', 'sobel_x', 'sobel_y', 'scharr_x']
fft_filters = [np.fft.fft2(x) for x in filters]
fft_shift = [np.fft.fftshift(y) for y in fft_filters]
mag_spectrum = [np.log(np.abs(z)+1) for z in fft_shift]

for i in range(len(filters)):
    plt.imshow(mag_spectrum[i], cmap='gray')
    plt.title(filter_name[i])
    plt.show()
```
